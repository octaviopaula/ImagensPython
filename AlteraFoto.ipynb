```python
from skimage import io
import numpy as np
import matplotlib .pyplot as plt
```

```python
image = io.imread('https://i.imgur.com/Jlz8aZj.png')
```

```python
plt.imshow(image)
```

```python
ImageGray = []
for line in image:
  lineGray = []
  for r, g, b in line:
    media = int(0.3 * r + 0.59 * g + 0.11 * b)
    lineGray.append(media)
  ImageGray.append(lineGray)
```

```python
plt.imshow(ImageGray, cmap="gray")
```

```python
R = 2
M = len(ImageGray) - R
N = len(ImageGray[0]) - R
```

```python
imageMediana = []
for i in range(R, M):
  newLine = []
  for j in range(R, N):
    filtro = []
    for x in range(i-R, i+R+1):
      for y in range(j-R, j+R+1):
        filtro.append(ImageGray[x][y])
    mediana = int(round(np.median(filtro)))
    newLine.append(mediana)
  imageMediana.append(newLine)
```

```python
plt.imshow(imageMediana, cmap="gray")
```

```python
M = len(imageMediana) - 1 
N = len(imageMediana[0]) - 1 

filtro = [[1, 0, -1],
 [2, 0, -2],
 [1, 0, -1]]


```

```python
imageSobelV = []
for i in range(1, M):
  newLine = []
  for j in range(1, N):
    newPixel =((filtro[0][0] * imageMediana[i-1][j-1])
    + (filtro[0][1] * imageMediana[i-1][j])
    + (filtro[0][2] * imageMediana[i-1][j+1])
    + (filtro[1][0] * imageMediana[i][j-1])
    + (filtro[1][1] * imageMediana[i][j])
    + (filtro[1][2] * imageMediana[i][j+1])
    + (filtro[2][0] * imageMediana[i+1][j-1])
    + (filtro[2][1] * imageMediana[i+1][j])
    + (filtro[2][2] * imageMediana[i+1][j+1]))
    newLine.append(abs(newPixel))
  imageSobelV.append(newLine)
```

```python
plt.imshow(imageSobelV, cmap="gray")
```

```python
M = len(imageMediana) - 1 
N = len(imageMediana[0]) - 1 

filtro = [[ 1, 2, 1],
 [ 0, 0, 0],
 [-1, -2, -1]]
```

```python
imageSobelH = []
for i in range(1, M):
  newLine = []
  for j in range(1, N):
    newPixel =((filtro[0][0] * imageMediana[i-1][j-1])
    + (filtro[0][1] * imageMediana[i-1][j])
    + (filtro[0][2] * imageMediana[i-1][j+1])
    + (filtro[1][0] * imageMediana[i][j-1])
    + (filtro[1][1] * imageMediana[i][j])
    + (filtro[1][2] * imageMediana[i][j+1])
    + (filtro[2][0] * imageMediana[i+1][j-1])
    + (filtro[2][1] * imageMediana[i+1][j])
    + (filtro[2][2] * imageMediana[i+1][j+1]))
    newLine.append(abs(newPixel))
  imageSobelH.append(newLine)
```

```python
plt.imshow(imageSobelH, cmap="gray")
```

```python
imagePrewitt = np.array(imageSobelV) + np.array(imageSobelH)
```

```python
plt.imshow(imagePrewitt, cmap="gray")
```
