# Histograma


```python
image = io.imread('https://i.imgur.com/6ndlZB9.jpg')

imageGray = []

for line in image:
  lineGray = []
  for r, g, b in line:
    cl = int(0.3*r + 0.59*g + 0.11*b)
    lineGray.append(cl)
  imageGray.append(lineGray)

plt.imshow(imageGray, cmap='gray')  
```

```python
histogram = np.zeros(256, dtype=int)

for line in imageGray:
  for c in line:
    histogram[c] += 1

print(sum(histogram))  
```
