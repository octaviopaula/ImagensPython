# Histogramas Acumulados


```python
image = io.imread('https://i.imgur.com/6ndlZB9.jpg')
```

```python
plt.imshow(image)
```

```python
imageGray = []

for line in image:
  lineGray = []
  for r, g, b in line:
    cl = int(0.3*r + 0.59*g + 0.11*b)
    lineGray.append(cl)
  imageGray.append(lineGray)

```

```python
plt.imshow(imageGray, cmap='gray') 
```

```python
histogram = np.zeros(256, dtype=int)

for line in image:
  for c in line:
    histogram[c] += 1

plt.bar(range(256), histogram) 
```

```python
cumulative = np.zeros(256, dtype=int)

cumulative[0] = histogram[0]
for i in range(1 ,256):
  cumulative[i] = histogram[i] + cumulative[i -1]

plt.bar(range(256), cumulative)  
```

```python
L = 256
M = len(imageGray)
N = len(imageGray[0])
termo = (L - 1) / (M * N)
```

```python
imageEq = []

for line in imageGray:
 newLine = []
 for c in line:
  newPixel = int(round(termo * cumulative[c]))
  newLine.append(newPixel)
imageEq.append(newLine)
```

```python
plt.imshow(imageEq , cmap='gray') 
```
