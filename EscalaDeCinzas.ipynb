# Combinação


```python
image = io.imread('https://i.imgur.com/6ndlZB9.jpg')
```

```python
plt.imshow(image)
```

```python
ImageGray = []
for line in image:
  lineGray = []
  for r, g, b in line:
    media = int(0.3 * r + 0.59 * g + 0.11 * b)
    lineGray.append(media)
  ImageGray.append(lineGray)
```
