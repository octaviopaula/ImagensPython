```python
from skimage import io
import numpy as np
import matplotlib .pyplot as plt
```

# Filtro de Média

```python
image = io.imread('https://i.imgur.com/DkUHiQT.gif')
```

```python
plt.imshow(image)
```

```python
histogram = np.zeros(256, dtype=int)

for line in image:
  for c in line:
    histogram[c] += 1

plt.bar(range(256), histogram)    
```

```python
cumulative = np.zeros(256, dtype=int)

cumulative[0] = histogram[0]
for i in range(1 ,256):
  cumulative[i] = histogram[i] + cumulative[i -1]

plt.bar(range(256), cumulative)  
```

```python
R = 2
M = len(image) - R
N = len(image[0]) - R
```

```python
imageMedia = []
for i in range(R, M):
 newLine = []
 for j in range(R, N):
  filtro = []
  for x in range(i-R, i+R+1):
    for y in range(j-R, j+R+1):
      filtro.append(image[x][y])
    media = int(round(np.mean(filtro)))
    newLine.append(media)
  imageMedia.append(newLine)
```

```python
plt.imshow(imageMedia)
```

# Filtro de Mediana



```python
image = io.imread('https://i.imgur.com/F6tXiCm.png')
```

```python
plt.imshow(image, cmap="gray")
```

```python
R = 1
M = len(image) - R
N = len(image[0]) - R
```

```python
imageMediana = []
for i in range(R, M):
  newLine = []
  for j in range(R, N):
    filtro = []
    for x in range(i-R, i+R+1):
      for y in range(j-R, j+R+1):
        filtro.append(image[x][y])
    mediana = int(round(np.median(filtro)))
    newLine.append(mediana)
  imageMediana.append(newLine)
```

```python
plt.imshow(imageMediana, cmap="gray")
```
